```python
df = pd.read_csv('hunan_other.csv') 
df
```

```python
import random

for index, row in df.iterrows():
    rd = random.uniform(0.2,0.25)
    df2 = pd.DataFrame([["Hunan",row['Latitude']+rd, row['Longitude']+1.5*rd]], columns=['province','Latitude','Longitude'])
    df = df.append(df2, ignore_index = True)

df

```

```python
import os
df.to_csv('hunanOthers.csv')
```
